# Nonlinear classifiers
* Some simple examples for nonlinear classification

```python
# Load files from introMLNB 
import sys
import os
targetDirectory = os.path.abspath("../../introML/")
sys.path.append(targetDirectory)

%matplotlib inline
import mglearn

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import colors

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
```

```python
# generate random points
# classify based on circle distance
def makeCircleClass(n):
    X = np.random.normal(size=(n,2),loc=0,scale=1.)
    d = X[:,0]**2. + X[:,1]**2.
    dtype = d > 1.5
    y = dtype.astype(int)
    return X,y
```

```python
X,y = makeCircleClass(100)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5)
```

```python
# draw points
width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.grid()
```

```python
# linear support vector classifier
svl = LinearSVC(C=100.,max_iter=10000)
trainFit = svl.fit(X_train,y_train)
print(trainFit.score(X_train,y_train))
print(trainFit.score(X_test,y_test))
print(np.mean(y_test))
```

```python
# plot linear split
width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
mglearn.plots.plot_2d_separator(trainFit,X, fill=False)
plt.grid()
```

```python
svnl = SVC(kernel='rbf',C=1.,gamma=1.0)
trainFit = svnl.fit(X_train,y_train)
print(trainFit.score(X_train,y_train))
print(trainFit.score(X_test,y_test))
```

```python
width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
mglearn.plots.plot_2d_separator(trainFit,X, fill=False)
plt.grid()
```

```python
svnl = SVC(kernel='poly',degree=2,C=1.,gamma=1.0)
trainFit = svnl.fit(X_train,y_train)
print(trainFit.score(X_train,y_train))
print(trainFit.score(X_test,y_test))
```

```python
width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
mglearn.plots.plot_2d_separator(trainFit,X, fill=False)
plt.grid()
```

# Multilayer Perceptron Classifier
* This applies the MLP classifier to our test data
* Note hidden unit controls
* Also, changes activiation
* alpha is a penalty on the weights, as in ridge regression


```python
from sklearn.neural_network import MLPClassifier
X,y = makeCircleClass(50)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5)
```

```python
mlp = MLPClassifier(solver='lbfgs',activation='relu',hidden_layer_sizes=[5])
#f（x）= max（0，x）
trainFit = mlp.fit(X_train,y_train)
print(trainFit.score(X_train,y_train))
print(trainFit.score(X_test,y_test))

width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
mglearn.plots.plot_2d_separator(trainFit,X, fill=False)
plt.grid()
```

```python
mlp = MLPClassifier(solver='lbfgs',activation='tanh',hidden_layer_sizes=[5])
trainFit = mlp.fit(X_train,y_train)
print(trainFit.score(X_train,y_train))
print(trainFit.score(X_test,y_test))

width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
mglearn.plots.plot_2d_separator(trainFit,X, fill=False)
plt.grid()
```

```python
mlp = MLPClassifier(solver='lbfgs',activation='relu',hidden_layer_sizes=[20,20])
trainFit = mlp.fit(X_train,y_train)
print(trainFit.score(X_train,y_train))
print(trainFit.score(X_test,y_test))

width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
mglearn.plots.plot_2d_separator(trainFit,X, fill=False)
plt.grid()
```

```python
mlp = MLPClassifier(solver='lbfgs',activation='relu',alpha=1.2,hidden_layer_sizes=[20,20])
trainFit = mlp.fit(X_train,y_train)
print(trainFit.score(X_train,y_train))
print(trainFit.score(X_test,y_test))

width = 10
height = 10
plt.figure(figsize=(width, height))
plt.rcParams.update({'font.size':15})
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape: {}".format(X.shape))
mglearn.plots.plot_2d_separator(trainFit,X, fill=False)
plt.grid()
```
